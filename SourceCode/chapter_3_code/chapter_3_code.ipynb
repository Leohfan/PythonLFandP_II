## 3.1.3 Python基本语句
### （1）表达式

```python
# 算术运算符
a = 10
b = 3

print(f"Addition: {a + b}")         # 加法，结果: 13
print(f"Subtraction: {a - b}")      # 减法，结果: 7
print(f"Multiplication: {a * b}")   # 乘法，结果: 30
print(f"Division: {a / b}")          # 除法，结果: 3.3333...
print(f"Floor Division: {a // b}")   # 取整除，结果: 3
print(f"Modulus: {a % b}")           # 取余，结果: 1
print(f"Exponentiation: {a ** b}")   # 幂运算，结果: 1000
```

```python
# 比较运算符
print(f"Equal: {a == b}")            # 等于，结果: False
print(f"Not Equal: {a != b}")        # 不等于，结果: True
print(f"Greater Than: {a > b}")      # 大于，结果: True
print(f"Less Than: {a < b}")         # 小于，结果: False
print(f"Greater or Equal: {a >= b}") # 大于等于，结果: True
print(f"Less or Equal: {a <= b}")    # 小于等于，结果: False
```

```python
# 逻辑运算符
x = True
y = False

print(f"Logical AND: {x and y}")     # 逻辑与，结果: False
print(f"Logical OR: {x or y}")       # 逻辑或，结果: True
print(f"Logical NOT: {not x}")       # 逻辑非，结果: False
```

```python
# 位运算符
c = 5  # 二进制 0101
d = 3  # 二进制 0011

print(f"Bitwise AND: {c & d}")       # 按位与，结果: 1 (二进制 0001)
print(f"Bitwise OR: {c | d}")        # 按位或，结果: 7 (二进制 0111)
print(f"Bitwise XOR: {c ^ d}")       # 按位异或，结果: 6 (二进制 0110)
print(f"Bitwise NOT: {~c}")          # 按位取反，结果: -6 (取反后为负数)
print(f"Left Shift: {c << 1}")       # 左移，结果: 10 (二进制 1010)
print(f"Right Shift: {c >> 1}")      # 右移，结果: 2 (二进制 0010)
```

```python
# 赋值运算符
e = 5
print(f"Initial Value: {e}")          # 初始值
e += 2  # 加法赋值
print(f"After += 2: {e}")             # 结果: 7
e -= 1  # 减法赋值
print(f"After -= 1: {e}")             # 结果: 6
e *= 3  # 乘法赋值
print(f"After *= 3: {e}")             # 结果: 18
e /= 2  # 除法赋值
print(f"After /= 2: {e}")             # 结果: 9.0
e //= 3 # 取整除赋值
print(f"After //= 3: {e}")            # 结果: 3.0
e %= 2  # 取余赋值
print(f"After %= 2: {e}")             # 结果: 1.0
e **= 3  # 幂赋值
print(f"After **= 3: {e}")            # 结果: 1.0
```

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a is b)        # 输出: False，因为a和b是不同的对象
print(a is c)        # 输出: True，因为a和c是同一个对象
print(a is not b)    # 输出: True，因为a和b不是同一个对象
```

```python
my_list = [1, 2, 3, 4, 5]
my_dict = {"name": "Alice", "age": 25}

print(3 in my_list)  # 输出: True，因为3在列表中
print(6 in my_list)  # 输出: False，因为6不在列表中
print("name" in my_dict)  # 输出: True，因为"name"是字典的键
print("Alice" not in my_dict)  # 输出: True，因为"Alice"不是字典的键
```

```python
# 定义两个集合
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

# 添加元素
set_a.add(6)  # 添加元素 6 到 set_a
print("After adding 6 to set_a:", set_a)
# 删除元素
set_a.remove(2)  # 删除元素 2 从 set_a
# set_a.discard(10)  # 如果要删除的元素不存在，remove() 会引发错误，而 discard() 不会引发错误
print("After removing 2 from set_a:", set_a)
# 交集
intersection = set_a & set_b  # 交集
print("Intersection of set_a and set_b:", intersection)
# 并集
union = set_a | set_b  # 并集
print("Union of set_a and set_b:", union)
# 差集
difference = set_a - set_b  # 差集
print("Difference of set_a and set_b (set_a - set_b):", difference)
# 对称差集
symmetric_difference = set_a ^ set_b  # 对称差集
print("Symmetric difference of set_a and set_b:", symmetric_difference)
```

```python
#生成一个包含平方数的列表：
squares = [x**2 for x in range(1, 6)]
print(squares)  # 输出: [1, 4, 9, 16, 25]

#过滤掉列表中的奇数：
even_squares = [x**2 for x in range(1, 6) if x % 2 == 0]
print(even_squares)  # 输出: [4, 16]

#将一个列表中的元素作为键，平方值作为值，生成字典：
numbers = [1, 2, 3, 4, 5]
square_dict = {x: x**2 for x in numbers}
print(square_dict)  # 输出: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

```python
#基于现有字典的值生成新的字典：
original_dict = {'a': 1, 'b': 2, 'c': 3}
new_dict = {k: v**2 for k, v in original_dict.items()}
print(new_dict)  # 输出: {'a': 1, 'b': 4, 'c': 9}

#生成1到10之间的偶数集合：
even_set = {x for x in range(1, 11) if x % 2 == 0}
print(even_set)  # 输出: {2, 4, 6, 8, 10}

#基于字符串生成字符集合（去除重复）：
chars = {char for char in 'hello world' if char != ' '}
print(chars)  # 输出: {'d', 'e', 'h', 'l', 'o', 'r', 'w'}
```

```python
gen = (x**2 for x in range(1, 6))
for val in gen:
    print(val)       # 输出: 1 4 9 16 25（每次迭代时计算生成）
```

```python
a = 10
b = 5

# 如果 a > b，返回 "a is greater"；否则返回 "b is greater or equal"
result = "a is greater" if a > b else "b is greater or equal"
print(result)  # 输出: a is greater
```

### 3.3.5 字符串格式化

```python
# 示例 1: 基本字符串格式化
name = "Alice"
age = 30
formatted_string = "My name is %s and I am %d years old." % (name, age)
print(formatted_string)
# 输出: My name is Alice and I am 30 years old.

# 示例 2: 使用标志和宽度
number = 7
formatted_number = "The number is %05d." % number  # 宽度为5，不足部分补零
print(formatted_number)
# 输出: The number is 00007.

# 示例 3: 使用精度
pi = 3.14159
formatted_pi = "Pi is approximately %.2f." % pi  # 小数点后保留两位
print(formatted_pi)
# 输出: Pi is approximately 3.14.

# 示例 4: 使用多个格式化选项
value = 123.456789
formatted_value = "Value: %10.3f" % value  # 宽度为10，小数点后保留三位
print(formatted_value)
# 输出: Value:     123.457
```

```python
# 使用空格作为填充符
print("{:_<10}".format("test"))  # 输出: test______
# 使用星号作为填充符
print("{:*^10}".format("test"))  # 输出: ***test***
# 左对齐
print("{:<10}".format("left"))   # 输出: left      
# 右对齐
print("{:>10}".format("right"))  # 输出:      right
# 居中对齐
print("{:^10}".format("center")) # 输出:   center  

# 设置宽度为10
print("{:10}".format("width"))  # 输出: width     
# 设置宽度为15，右对齐
print("{:>15}".format("Python"))  # 输出:         Python

# 使用逗号分隔数字
print("{:,}".format(1234567890))  # 输出: 1,234,567,890
# 对浮点数使用逗号
print("{:,.2f}".format(1234567.8910))  # 输出: 1,234,567.89
# 保留两位小数
print("{:.2f}".format(3.14159))  # 输出: 3.14
# 保留四位小数
print("{:.4f}".format(2.71828))  # 输出: 2.7183
```

```python
# 字符串
print("{:s}".format("Hello"))  # 输出: Hello
# 十进制整数
print("{:d}".format(123))      # 输出: 123
# 八进制整数
print("{:o}".format(123))      # 输出: 173
# 十六进制整数
print("{:x}".format(123))      # 输出: 7b
# 科学计数法
print("{:e}".format(123.456))  # 输出: 1.234560e+02

# 组合使用：输出一个右对齐、宽度为15的带千位分隔符的浮点数，保留两位小数
print("{:*>15,.2f}".format(12345.6789))  # 输出: *****12,345.68
# 组合使用：居中对齐，填充字符为#，宽度为12，显示浮点数保留4位小数
print("{:#^12.4f}".format(3.14159))      # 输出: ###3.1416###
# 组合使用：字符串左对齐，宽度为15，填充字符为空格
print("{:<15}".format("Python"))         # 输出: Python         
```

```python
products = [("苹果", 1.2, 10), ("香蕉", 0.5, 20), ("橙子", 0.8, 15)]
print(f"{'商品':<10} {'单价':<10} {'数量':<10} {'总价':<10}")
for product in products:
    name, price, quantity = product
    total_price = price * quantity
    print(f"{name:<10} {price:<10.2f} {quantity:<10} {total_price:<10.2f}")
```
