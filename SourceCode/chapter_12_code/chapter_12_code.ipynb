## 12.5 多图绘制

### 12.5.1 使用 subplots 绘制多个子图

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建 2 行 2 列的子图网格
fig, axes = plt.subplots(2, 2, figsize=(8, 4))
# 在每个子图中绘制不同的数据
x = np.linspace(0, 10, 100)

# 第一个子图
axes[0,0].plot(x, np.sin(x), label="sin(x)", color='blue')
axes[0,0].set_title("子图 1: sin(x)")
axes[0,0].legend()

# 第二个子图
axes[0,1].plot(x, np.cos(x), label="cos(x)", color='green')
axes[0,1].set_title("子图 2: cos(x)")
axes[0,1].legend()

# 第三个子图
axes[1,0].plot(x, np.tan(x), label="tan(x)", color='red')
axes[1,0].set_title("子图 3: tan(x)")
axes[1,0].legend()

# 第四个子图
axes[1,1].plot(x, np.exp(-x), label="exp(-x)", color='purple')
axes[1,1].set_title("子图 4: exp(-x)")
axes[1,1].legend()

plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定字体为黑体
plt.rcParams['axes.unicode_minus'] = False    # 解决负号无法显示的问题

plt.tight_layout()  # 自动调整布局避免重叠
plt.show()          # 显示图表
```

### 12.5.3 网格布局

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import numpy as np

# 创建网格布局
fig = plt.figure(figsize=(8, 4))
gs = GridSpec(3, 3, figure=fig)   # 创建一个 3×3 的网格。

# 在网格中添加子图
ax1 = fig.add_subplot(gs[0, :])   # 占据第一行的所有列
ax2 = fig.add_subplot(gs[1, :-1]) # 占据第二行的前两列
ax3 = fig.add_subplot(gs[1:, -1]) # 占据第二行和第三行的最后一列
ax4 = fig.add_subplot(gs[2, :-1]) # 占据第三行的前两列

# 绘制数据
x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x), label="sin(x)")
ax1.set_title("子图 1")

ax2.plot(x, np.cos(x), label="cos(x)", color='orange')
ax2.set_title("子图 2")

ax3.plot(x, np.tan(x), label="tan(x)", color='red')
ax3.set_title("子图 3")

ax4.plot(x, np.exp(-x), label="exp(-x)", color='green')
ax4.set_title("子图 4")

plt.tight_layout()
plt.show()
```

## 12.6 综合应用案例

社交媒体广告投放效果的综合评估
	
	描述：以某品牌在多个社交媒体平台的广告投放效果数据为背景，结合堆叠柱状图、散点图和子图布局，展示广告投入、点击量、转化率和平台受众分布的综合分析。要求体现投入产出比，并通过图表直观揭示最佳的广告投放策略。

```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
platforms = ["Facebook", "Instagram", "Twitter", "LinkedIn", "TikTok"]
ad_spend = [5000, 2500, 1500, 4000, 3500]  # 广告投入 ($)
clicks = [1500, 900, 750, 500, 800]   # 点击量

# 转化量
conversions = [100, 80, 120, 30, 60]
# 转化率 (%)
conversion_rate = [c / cl * 100 if cl > 0 else 0 for c, cl in zip(conversions, clicks)] 
# 平台受众分布 (%)
audience_distribution = [30, 25, 20, 15, 10]  

# 创建多图布局
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Social Media Ad Performance Analysis", fontsize=16)

# 1. 堆叠柱状图：展示广告投入与点击量的对比
bar_width = 0.5
x = np.arange(len(platforms))

axs[0,0].bar(x, ad_spend, width=bar_width, 
                label="Ad Spend ($)", color="skyblue")
axs[0,0].bar(x, clicks, width=bar_width, bottom=ad_spend, 
                label="Clicks", color="orange")
axs[0,0].set_title("Ad Spend vs Clicks")
axs[0,0].set_xticks(x)
axs[0,0].set_xticklabels(platforms)
axs[0,0].set_ylabel("Value")
axs[0,0].legend()

# 2. 散点图：展示广告投入与转化量的关系
axs[0,1].scatter(ad_spend, conversions, color="purple", s=100, alpha=0.7)
for i, txt in enumerate(platforms):
    axs[0,1].annotate(txt, (ad_spend[i], conversions[i]), fontsize=10)

axs[0,1].set_title("Ad Spend vs Conversions")
axs[0,1].set_xlabel("Ad Spend ($)")
axs[0,1].set_ylabel("Conversions")
axs[0,1].grid(True)

# 3. 条形图：展示转化率
axs[1,0].bar(platforms, conversion_rate, color="green", alpha=0.7)
axs[1,0].set_title("Conversion Rate by Platform")
axs[1,0].set_ylabel("Conversion Rate (%)")
axs[1,0].set_xlabel("Platforms")
axs[1,0].grid(axis="y")

# 4. 饼图：展示平台受众分布
axs[1,1].pie(audience_distribution, 
                labels=platforms, 
                autopct="%1.1f%%", 
                startangle=90, 
                colors=["lightblue", "pink", "lightgreen", "yellow", "coral"])
axs[1,1].set_title("Audience Distribution by Platform")

# 调整布局
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

	- 产品 B：\\([180, 200, 230, 220, 210, 240, 250, 260, 270, 280, 290, 300]\\)
	- 产品 C：\\([100, 150, 200, 180, 170, 190, 200, 210, 220, 230, 240, 250]\\)
	- 产品 D：\\([150, 180, 210, 200, 190, 220, 230, 240, 250, 260, 270, 280]\\)

1. 绘制产品月销售量的并列柱状图，进行对此分析。

```python
import matplotlib.pyplot as plt
import numpy as np
# 数据准备
month = np.arange(1, 13)  # 1 到 12 月份
width = 0.2  # 柱状图宽度
product_A = np.array([200, 250, 300, 270, 220, 310, 260, 280, 300, 340, 310, 330])
product_B = np.array([180, 200, 230, 220, 210, 240, 250, 260, 270, 280, 290, 300])
product_C = np.array([100, 150, 200, 180, 170, 190, 200, 210, 220, 230, 240, 250])
product_D = np.array([150, 180, 210, 200, 190, 220, 230, 240, 250, 260, 270, 280])

# 绘制并列柱状图
plt.bar(month-1.5*width,product_A,width=width,label='Product A',color='skyblue')
plt.bar(month-0.5*width,product_B,width=width,label='Product B',color='orange')
plt.bar(month +0.5*width,product_C,width=width,label='Product C',color='green')
plt.bar(month +1.5*width,product_D,width=width,label='Product D',color='red')

# 添加图表标题和轴标签, 显示图表
plt.title("Monthly Sales Comparison of Products")
plt.xlabel("Months")
plt.ylabel("销售额 (in $1000)")
plt.xticks(month, labels=[f"{i}月" for i in month])  # 设置 x 轴刻度
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

2. 绘制产品月销售量的堆叠柱状图，观察月销售总量的分布。

```python
# 计算每个阶段的堆叠数据
product_B_bottom = product_A
product_C_bottom = product_A + product_B
product_D_bottom = product_A + product_B + product_C

# 绘制堆叠柱状图
plt.bar(month, product_A, label='Product A')
plt.bar(month, product_B, bottom=product_B_bottom, label='Product B')
plt.bar(month, product_C, bottom=product_C_bottom, label='Product C')
plt.bar(month, product_D, bottom=product_D_bottom, label='Product D')

plt.title("Monthly Sales of Products")
```

3. 绘制折线图，展示每个产品的月度销售趋势。

```python
# 绘制每个产品的月度销售趋势
plt.plot(month,product_A,label='Product A',marker='o',linestyle='-',color='blue')
plt.plot(month,product_B,label='Product B',marker='x',linestyle='--',color='green')
plt.plot(month,product_C,label='Product C',marker='s',linestyle='-.',color='red')
plt.plot(month,product_D,label='Product D',marker='d',linestyle=':',color='purple')

plt.title("Monthly Sales Trend of Products")
```

4. 计算每个产品的市场份额，绘制饼图，展示各个产品在总销售中的占比。

```python
# 计算每个产品的总销售额
total_sales_A = product_A.sum()
total_sales_B = product_B.sum()
total_sales_C = product_C.sum()
total_sales_D = product_D.sum()

# 计算各产品的市场份额
total_sales = total_sales_A + total_sales_B + total_sales_C + total_sales_D
market_share_A = total_sales_A / total_sales * 100
market_share_B = total_sales_B / total_sales * 100
market_share_C = total_sales_C / total_sales * 100
market_share_D = total_sales_D / total_sales * 100

# 绘制饼图
labels = ['Product A', 'Product B', 'Product C', 'Product D']
sizes = [market_share_A, market_share_B, market_share_C, market_share_D]
colors = ['blue', 'green', 'red', 'purple']
explode = (0.1, 0, 0, 0)  # 突出显示第一个切片（Product A）

plt.pie(sizes, labels=labels, colors=colors, explode=explode,
        autopct='%1.1f%%', startangle=140)
plt.title("Market Share of Products")
```

(1) 使用`Pandas`查看数据集整体数据，并作描述性统计；

(2) 编写程序绘制纵轴表示"sepal length"、"sepal width"、"petal length"和"petal width"的折线图，并已不同颜色、不同线型区分；

(3) 编写程序将数据集上的每个特征属性取值分成6个区间，并绘制其对应的柱状图，其中柱状图横轴表示区间编号，纵轴表示每个区间对应的样本数量；（提示：区间划分时注意不应超过每个特征属性的最大值与最小值）；

(4) 编写程序绘制以"sepal length"为$x$轴、"sepal width"为$y$轴的散点图，以及"petal length"为$x$轴、"petal width"为$y$轴的散点图，并在图中用不同颜色、不同散点类型区分出鸢尾花类别（`Iris Setosa`—红色`o`、`Iris Versicolour`—蓝色`*`、`Iris Virginica`—绿色`+`）。

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv("irisdata.csv")
df.head()
```

```python
df.dtypes
```

```python
df["class"].value_counts()
```

```python
df.describe()
```

```python
plt.plot(range(1,151),df["sepal length in cm"],color='g',ls='-.',label='sepal length in cm')  
plt.plot(range(1,151),df["sepal width in cm"],color='r',label='sepal width in cm')
plt.plot(range(1,151),df["petal length in cm"],color='b',ls='--',label='petal length in cm ')
plt.plot(range(1,151),df["petal width in cm"],color='k',ls=':',label='petal width in cm')
plt.xlabel("id")
plt.ylabel("length/width in cm")
plt.legend()
plt.show()
```

```python
sl_cut = pd.cut(df["sepal length in cm"], bins =[df["sepal length in cm"].min(),4.9,5.5,6.1,6.6,7.2,df["sepal length in cm"].max()])
sl_cut.value_counts()
```

```python
sw_cut=pd.cut(df["sepal width in cm"],bins=[df["sepal width in cm"].min(),2.4,2.8,3.2,3.6,4.0,df["sepal width in cm"].max()])
sw_cut.value_counts()
```

```python
pl_cut = pd.cut(df["petal length in cm"], bins =[df["petal length in cm"].min(),1.9,2.8,3.7,4.6,5.5,df["petal length in cm"].max()])
pl_cut.value_counts()
```

```python
pw_cut=pd.cut(df["petal width in cm"],bins=[df["petal width in cm"].min(),0.5,0.9,1.3,1.7,2.1,df["petal width in cm"].max()])
pw_cut.value_counts()
```

```python
X1 = np.arange(len(sl_cut.value_counts()))
Y1 = sl_cut.value_counts()
Y2 = sw_cut.value_counts()
Y3 = pl_cut.value_counts()
Y4 = pw_cut.value_counts()
plt.bar(X1,Y1,color='r',width=0.2,label="sl")
plt.bar(X1+0.2,Y2,color='g',width=0.3,label="sw")
plt.bar(X1+0.2*2,Y3,color='b',width=0.3,label="pl")
plt.bar(X1+0.2*3,Y4,color='k',width=0.3,label="pw")
plt.xlabel("Group No.")
plt.ylabel("Number")
plt.legend()
plt.show()
```

```python
y=df["class"]
plt.scatter(df["sepal length in cm"][y=="Iris-setosa"],df["sepal width in cm"][y=="Iris-setosa"],color='r',marker='o') 
plt.scatter(df["sepal length in cm"][y=="Iris-versicolor"],df["sepal width in cm"][y=="Iris-versicolor"],color='b',marker='*')
plt.scatter(df["sepal length in cm"][y=="Iris-virginica"],df["sepal width in cm"][y=="Iris-virginica"],color='g',marker='+')
plt.xlabel("sepal length in cm")
plt.ylabel("sepal width in cm")
plt.grid()          #显示网格线
plt.show()
```

```python
y=df["class"]
plt.scatter(df["petal length in cm"][y=="Iris-setosa"],df["petal width in cm"][y=="Iris-setosa"],color='r',marker='o') 
plt.scatter(df["petal length in cm"][y=="Iris-versicolor"],df["petal width in cm"][y=="Iris-versicolor"],color='b',marker='*')
plt.scatter(df["petal length in cm"][y=="Iris-virginica"],df["petal width in cm"][y=="Iris-virginica"],color='g',marker='+')
plt.xlabel("petal length in cm")
plt.ylabel("petal width in cm")
plt.grid()       #显示网格线
plt.show()
```
