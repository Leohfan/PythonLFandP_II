### 4.3.6 选择结构的应用

```python
import time

# 获取当前时间的本地时间信息
date = time.localtime()
year, month, day = date[:3]  # 提取当前的年、月、日

# 列出每个月的天数，默认情况下2月为28天
day_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

# 判断是否为闰年
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    day_month[1] = 29  # 闰年时2月份的天数为29天

# 计算当前日期是当年的第几天
if month == 1:
    sum_day = day  # 如果是1月份，当天的日期就是全年的天数
else:
    # 对于其他月份，计算之前每月的天数与当前日期
    sum_day = sum(day_month[:month - 1]) + day  

# 输出结果
print('Today is %d.%d.%d, it is the %dth day of %d.' % (year, month, day, sum_day, year))
```

## 4.4 循环结构

### 4.4.1 while循环

```python
# 初始化累加器
total_sum = 0
# 初始化计数器
number = 1

# 使用while循环进行累加
while number <= 100:
    total_sum += number  # 将当前数字加入总和
    number += 1          # 计数器自增1

# 输出结果
print("The sum of integers from 1 to 100 is:", total_sum)
```

```python
# 定义要输出的字符串, 初始化索引
text, index = "Python", 0

# 使用 while-else 循环逐个输出字符
while index < len(text):
    if index == len(text) - 1:  # 如果是最后一个字符
        print(text[index])      # 直接输出最后一个字符后换行
    else:
        print(text[index], end=', ')  # 否则输出字符并添加逗号
    index += 1  # 索引自增1
else:
    # 当 while 循环正常结束后执行
    print("程序结束！")
```

### 4.4.2 for循环

```python
# 初始化累加和
sum_total = 0

# 使用 for 循环遍历1到100的所有整数
for i in range(1, 101):
    sum_total += i  # 将当前整数累加到总和中

# 输出结果
print("1到100之间所有整数的累加之和为:", sum_total)
```

```python
# 定义要输出的字符串
text = "Python"

# 使用 for 循环逐个输出字符
for char in text:
    print(char, end=', ')  # 输出当前字符，后面加逗号
else:
    # 当 for 循环正常结束后执行
    print("\n程序结束！")  # 输出提示信息
```

```python
# 输出九九乘法表
for i in range(1, 10):  # 外层循环，i 从 1 到 9
    for j in range(1, i + 1):  # 内层循环，j 从 1 到 i
        print(f"{j}x{i}={j * i}", end="\t")  # 输出乘法结果，使用制表符进行分隔
    print()  # 换行，开始新的一行
else:
    print('Program is done.')
```

### 4.4.3 break语句

```python
# 使用 for 循环遍历字符串
for char in 'Python':
    if char == 'h':
        break  # 当遇到字母 'h' 时，终止循环
    print('当前字母:', char)  # 输出当前字符
```

```python
#break打印九九乘法表
for i in range(1, 10):  # 外层循环，i 从 1 到 9
    for j in range(1, i + 1):  # 内层循环，j 从 1 到 i
        if j == 5:      # 当 j 等于 5 时终止内层循环
            break
        print(f"{j}x{i}={j * i}", end="\t")  # 输出乘法结果
    print('第', i, '行')  # 输出当前行的结束
```

```python
# 输出1到50之间的素数
for num in range(2, 51):  # 从2到50（包括50）
    is_prime = True       # 假设num是素数
    for i in range(2, int(num**0.5) + 1):  # 检查从2到num的平方根
        if num % i == 0:  # 如果能被i整除
            is_prime = False  # 不是素数
            break  # 跳出内层循环
    if is_prime:   # 如果是素数
        print(num, end=" ")  # 输出素数，所有素数在一行中输出
else:   # 当 for 循环执行完毕时，执行 else 语句块    
    print("\nProgram is done.")
```

### 4.4.3 continue语句

```python
# 使用 for 循环遍历字符串
for char in 'Python':
    if char == 'h':
        continue  # 当遇到字母 'h' 时，中止本次循环
    print('当前字母:', char)  # 输出当前字符
```

```python
#continue打印九九乘法表
for i in range(1, 10):
    for j in range(1, i + 1):
        if (i * j) % 6 == 0:
            continue     # 跳过乘积为6的倍数
        print(f"{j}x{i}={i * j}", end="\t")
    print('第', i, '行')  # 输出当前是第几行
```

```python
#异常处理try-except语句
try:
    a = 100
    b = int(input("请输入一个数："))
    print(a/b)
except(ZeroDivisionError):
    print("输入的数据不符合要求，程序出错！")
```

```python
while True:
    try:
        str = "0123456789ABCDEF"
        n = int(input("请输入一个整数："))
        print('十六进制数字的第%d个是%s.'%(n,str[n]))
        break
    except (ValueError,IndexError):
        print("输入错误，请输入一个合适的整数！")
```

```python
a = 100
try:
    b = eval(input("请输入一个数："))
    print(a/b)
except (ZeroDivisionError) as e:
    print(f"发生了异常: {e}")
```

```python
#try-except-else语句
try:
    a = 100
    b = eval(input("请输入一个数："))
    print('%d/%d = %.2f'%(a,b,a/b))
except(ZeroDivisionError) as e:
    print(e)
else:
    print('程序运行完成')
```

```python
#try-except-else-finally语句
while True:
    try:
        # 定义十六进制字符串
        hex_digits = "0123456789ABCDEF"
        # 接受用户输入并转换为整数
        n = int(input("请输入一个整数（0-15）："))

        # 检查输入范围
        if n < 0 or n > 15:
            raise IndexError("输入的整数超出范围！")

        # 输出对应位置的十六进制数字
        print('十六进制数字的第%d个是%s.'%(n, hex_digits[n]))
```

```python
#第4题
def division():
    apple = int(input("请输入苹果的个数:"))
    children = int(input("请输入小朋友个数:"))
    result = apple // children
    remain = apple - result * children
    if remain > 0:
        print(apple,"个苹果，平均分给",children,"个小朋友，每人分",result,"个，剩下",remain,"个。")
    else:
        print(apple,"个苹果，平均分给",children,"个小朋友，每人分",result,"个。")
while True:
    try:
        division()
    except ZeroDivisionError:
        print("出错了，不能有0个小朋友，请重新输入!")
    except ValueError:
        print("输入错误，请重新输入!")
    else:
        print("苹果顺利分完!")
        break
```
