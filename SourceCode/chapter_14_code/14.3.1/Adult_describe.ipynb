```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import collections

#
raw_data = pd.read_csv('Adult.csv',na_values=' ?')   #' ?'
fill_na = lambda col:col.fillna(col.mode()[0])   #fill_na
fill_data = raw_data.apply(fill_na, axis = 0)   #fill_data
data = fill_data.drop(['fnlwgt'],axis=1)   #'fnlwgt'

#
from collections import Counter
items = data['workclass']   #
count = Counter(items)   #
print(count)   #
countSort = collections.OrderedDict()
countSort = dict(count.most_common())

#
plt.rcParams['figure.dpi'] = 300 #

#
sizes = list(countSort.values())   
for i in range(len(sizes)):
    sizes[i] = round(sizes[i]/32561 * 100,2)
labels = list(countSort.keys())   #
plt.pie(sizes,labels=labels,autopct='%.2f',counterclock=False,startangle = 0)   #
plt.savefig('Distribution of workclass.jpg')
plt.show()
```
